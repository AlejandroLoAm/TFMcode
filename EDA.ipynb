```python
import numpy as np
import pandas as pd
import json

def transform_column_to_encoded_features(df, column, classes, prefix_classes):
    """
    Transforms a DataFrame column containing JSON lists into multiple feature columns.

    Parameters:
    - df (pd.DataFrame): The input DataFrame.
    - column (str): The column to process (containing JSON-like lists).
    - classes (list): The list of class names corresponding to the feature dimensions.
    - prefix_classes (str): The prefix to apply to the new column names.

    Returns:
    - pd.DataFrame: The DataFrame with the original column replaced by the new feature columns.
    """
    # Generate prefixed column names
    prefixed_classes = [f"{prefix_classes}_{cls}" for cls in classes]

     # Step 1: Parse JSON strings into Python lists and calculate max for each row
    def process_row(json_data):
        parsed_data = json.loads(json_data)  # Parse JSON string into a Python object (list of lists)
        array = np.array(parsed_data)  # Convert to a 2D NumPy array (21 x 56)
        max_values = np.max(array, axis=0)  # Calculate max along axis 0
        return max_values

    # Apply the process_row function to each row in the specified column
    max_moods = df[column].apply(process_row).tolist()

    # Step 2: Create a DataFrame with the calculated max values
    feature_df = pd.DataFrame(max_moods, columns=prefixed_classes)

    # Step 3: Drop the original column and concatenate the new features
    df = df.drop(columns=[column]).reset_index(drop=True)
    df = pd.concat([df, feature_df], axis=1)

    return df
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Extract identifiers
identifiers = essentia_genre[['song_id', 'title', 'artist_name']]

# Select numerical features (excluding identifiers)
features = essentia_genre.drop(columns=['song_id', 'title', 'artist_name']).values

# Step 1: Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 2: Perform PCA with 40 components
pca = PCA(n_components=40)
pca_features = pca.fit_transform(scaled_features)

# Check explained variance ratio
print("Explained variance by each component:", pca.explained_variance_ratio_)
print("Total explained variance:", np.sum(pca.explained_variance_ratio_))

# Convert PCA components to a DataFrame
pca_columns = [f"PC{i+1}" for i in range(40)]
pca_df = pd.DataFrame(pca_features, columns=pca_columns)
```

```python
# Combine identifiers with PCA components
result_df = pd.concat([identifiers.reset_index(drop=True), pca_df], axis=1)

# Display the resulting DataFrame
print(result_df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot explained variance ratio for each component
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.xlabel('PCA Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio of Each PCA Component')
plt.show()

# Cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.axhline(y=0.9, color='r', linestyle='--', label='90% Variance')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# Dimensión de entrada (40 componentes PCA)
input_dim = pca_features.shape[1]
encoding_dim = 20  # Tamaño de la capa de cuello de botella

# Definir la arquitectura del autoencoder
input_layer = Input(shape=(input_dim,))
encoded = Dense(30, activation="relu")(input_layer)  
bottleneck = Dense(encoding_dim, activation="relu", name="bottleneck")(encoded)  
decoded = Dense(30, activation="relu")(bottleneck)  
output_layer = Dense(input_dim, activation="linear")(decoded)  

# Compilación
autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer="adam", loss="mse")
autoencoder.summary()
```

```python
# Entrenar el autoencoder
history = autoencoder.fit(
    pca_features, 
    pca_features, 
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    shuffle=True,
    verbose=1
)
```

```python
# Extraer el encoder
encoder = Model(inputs=input_layer, outputs=bottleneck)

# Obtener la representación comprimida
compressed_features = encoder.predict(pca_features)

# Convertir a DataFrame
compressed_df = pd.DataFrame(compressed_features, columns=[f"Encoded_{i+1}" for i in range(encoding_dim)])

# Combinar con identificadores
final_df = pd.concat([result_df[['song_id','title', 'artist_name']].reset_index(drop=True), compressed_df], axis=1)
print(final_df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Extraer características comprimidas
compressed_features = final_df[[col for col in final_df.columns if col.startswith("Encoded_")]].values
identifiers = final_df[['song_id','title', 'artist_name']]

# Calcular la similitud
similarity_matrix = cosine_similarity(compressed_features)

# Convertir a DataFrame
similarity_df = pd.DataFrame(similarity_matrix, index=identifiers['song_id'], columns=identifiers['song_id'])
```

```python
def recommend_similar_songs(song_id, similarity_df, identifiers, top_n=5):
    """
    Recommends songs similar to the given song id.
    
    Parameters:
    - song_id (str): The id of the song for which recommendations are made.
    - similarity_df (pd.DataFrame): Cosine similarity matrix as a DataFrame.
    - identifiers (pd.DataFrame): DataFrame containing 'song_id' and 'artist_name'.
    - top_n (int): Number of recommendations to return.
    
    Returns:
    - pd.DataFrame: Recommended songs with ids and artist names.
    """
    if song_id not in similarity_df.index:
        return f"Song '{song_id}' not found in the database."
    
    # Get similarity scores for the given song
    song_scores = similarity_df[song_id].drop(index=[song_id])
    
    # Sort songs by similarity score in descending order
    top_similar_indices = song_scores.sort_values(ascending=False).iloc[:top_n].index
    
    # Retrieve song ids and artist names
    recommendations = identifiers[identifiers['song_id'].isin(top_similar_indices)]
    
    return recommendations
```

```python
def evaluate_recommendations(recommendations, user_id, filtered_interactions):
    
    # Convertir las recomendaciones en una lista de song_id
    recommended_song_ids = recommendations["song_id"].tolist()
    
    # Paso 2: Filtrar las interacciones del usuario para las canciones recomendadas
    user_interactions = filtered_interactions[filtered_interactions['user_id'] == user_id]
    
    # Paso 3: Verificar si el usuario ha escuchado alguna de las canciones recomendadas
    listened_songs = user_interactions[user_interactions['song_id'].isin(recommended_song_ids)]
    
    # Paso 4: Evaluar cuántas canciones recomendadas el usuario ha escuchado y cuántas veces
    songs_listened_count = listened_songs['song_id'].nunique()
    total_play_count = listened_songs['play_count'].sum()
    
    # Resultado
    result = {
        'user_id': user_id,
        'songs_listened_count': songs_listened_count,
        'total_play_count': total_play_count,
        'recommended_song_ids': recommended_song_ids
    }
    
    return result
```

```python
def evaluate_recommendations_for_all_songs_optimized(filtered_interactions, similarity_df, identifiers):
    from tqdm import tqdm
    results = []     

    for song_id in tqdm(filtered_interactions['song_id'].unique(), desc="Processing songs", unit="song"):
         
        users_for_song = filtered_interactions[filtered_interactions['song_id'] == song_id]['user_id'].unique()

        interactions_for_song = filtered_interactions[
            filtered_interactions['user_id'].isin(users_for_song)
        ]

        recommendations = recommend_similar_songs(song_id, similarity_df, identifiers)
        recommended_song_ids = recommendations["song_id"].tolist()

        interactions_for_recommended_songs = interactions_for_song[
            interactions_for_song['song_id'].isin(recommended_song_ids)
        ]
        
        song_play_counts = interactions_for_recommended_songs.groupby('song_id').agg(
            total_play_count=('play_count', 'sum'),
            success_recomendation=('user_id', 'nunique')
        ).reset_index()

        song_ids = song_play_counts['song_id'].values
        total_play_counts = song_play_counts['total_play_count'].values
        success_recomendation = song_play_counts['success_recomendation'].values

        for idx in range(len(song_ids)):
            result = {
                'original_song_id': song_id,
                'recommended_song_id': song_ids[idx],
                'success_recomendation': success_recomendation[idx],
                'total_play_count': total_play_counts[idx],
            }
            results.append(result)
    
    # Convertir los resultados a un DataFrame
    results_df = pd.DataFrame(results)
    
    return results_df

# Ejemplo de uso
results_df = evaluate_recommendations_for_all_songs_optimized(filtered_interactions, similarity_df, identifiers)
```
